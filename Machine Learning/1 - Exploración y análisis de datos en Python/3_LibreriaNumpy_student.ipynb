Hay un número de operaciones que puedes hacer con arrays numpy. Una de las más importantes es **slicing**, que consiste en extraer un subarray de un array.

## 2. Definiciones de vectores con un rango de valores

A continuación fíjate en las siguientes definiciones de vectores. Son muy útiles en ML o cualquier problema en el que queramos explorar un conjunto de valores equiespaciados o según escala logarítmica.


```python
print(np.arange(0, 100, 2))
```

El primer parámetro (0) es el punto inicial, el segundo (100) es el punto final sin incluirlo y el último (2) es el salto entre elementos.

```python
print(np.logspace(-6, 6, 13))
```

El primer parámetro (-6) es el exponente del punto inicial (10 elevado a -6), el segundo (6) es el exponente del punto final (10 elevado a 6) y el último (13) es el número de elementos.

## 3. Operaciones con matrices (EXTRA!)
### 3.1. Suma de matrices

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A)
B = np.array([[1, 2, 3], [4, 5, 6]])
print(B)
C = A + B
print(C)
```

### 3.2. Resta de matrices

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A)
B = np.array([[0.5, 0.5, 0.5], [0.5, 0.5, 0.5]])
print(B)
C = A - B
print(C)
```

### 3.3. Multiplicación de matrices (element-wise)

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A)
B = np.array([[1, 2, 3], [4, 5, 6]])
print(B)
C = A * B
print(C)
```

### 3.4. División de matrices

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A)
B = np.array([[1, 2, 3], [4, 5, 6]])
print(B)
C = A / B
print(C)
```

### 3.5. Producto escalar de matrices

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
print(A)
B = np.array([[1, 2], [3, 4]])
print(B)
C = A.dot(B)
print(C)
```

### 3.6. Producto de matrices y vectores

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
print(A)
B = np.array([0.5, 0.5])
print(B)
C = A.dot(B)
print(C)
D = A @ B
print(D)
```

### 3.7. Producto de matrices y escalares

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
print(A)
b = 0.5
print(b)
C = A * b
print(C)
```

### 3.8. Matriz transpuesta

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
print(A)
C = A.T
print(C)
```

### 3.9. Matriz inversa

```python
from numpy.linalg import inv
# define matrix
A = np.array([[1.0, 2.0], [3.0, 4.0]])
print(A)
# invert matrix
B = inv(A)
print(B)
# multiply A and B
I = A.dot(B)
print(I)
```
